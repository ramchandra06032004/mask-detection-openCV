```python
from tensorflow.keras.models import load_model

# Load the saved model
model = load_model('bestModel.keras')
```

```python
def detect_face_mask(img):
    y_pred=model.predict(img.reshape(1,224,224,3))
    return y_pred
```

```python
img=cv2.imread("1.png")
img=cv2.resize(img, (224, 224))
y_pred=detect_face_mask(img)
finalAns=np.argmax(y_pred)
print(finalAns)
```

```python
cap=cv2.VideoCapture(0)
```

```python
import cv2
import numpy as np

# Load the Haar Cascade for frontal face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame")
        break

    # Convert frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Crop the face from the frame
        face = frame[y:y+h, x:x+w]
        
        # Resize the face to the input size of the model
        img = cv2.resize(face, (224, 224))
        
        # Pass the cropped face to the model
        y_pred = detect_face_mask(img)
        finalAns = np.argmax(y_pred)
        
        # Define labels and colors
        labels = ["Incorrect Mask", "Mask", "No Mask"]
        colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]  # Blue, Green, Red

        # Get the label and color for the prediction
        label = labels[finalAns]
        color = colors[finalAns]

        # Display the label on the frame
        cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2, cv2.LINE_AA)
        cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)

    cv2.imshow("Frame", frame)
    print(label)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```
